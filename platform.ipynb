```python
games_path = "csvfiles/game.csv"
games_data = pd.read_csv(games_path, index_col=False)
```

```python
games_data.describe()
```

```python
# games_data.drop(columns="Unnamed: 0")
games_data.head()
```

```python
games_data["max rating"].max()
```

```python
#Get relative rating
relative_rating = games_data["rating"]/games_data["max rating"]
games_data["relative rating"] = relative_rating
# relative_rating = relative_rating * 5
games_data["relative rating"] = games_data["relative rating"] * 5
games_data
# # normalise rating score on the same base of 5
games_df = games_data.loc[games_data["relative rating"] <= 5]
games_df.describe()
```

```python
# Get the average number of platforms based on games ratings
average_platforms_df = games_data.groupby(["rating"]).mean()
average_platforms_df
```

```python
# Make Mouse ID a column instead of index
average_platforms_df["rating"] = average_platforms_df.index

# Reset index and reorder columns
average_platforms_df = average_platforms_df.reset_index(drop=True)
platforms_rating_df = average_platforms_df[["rating","number of platforms"]]

platforms_rating_df.head()
```

```python
plt.scatter(platforms_rating_df["rating"], platforms_rating_df["number of platforms"], facecolors="red", edgecolors="black") 
plt.title("Game ratings vs Average number of Stores")
plt.xlabel("Ratings score")
plt.ylabel("Average number of platforms")
```

```python
# Calculate the correlation coefficient and linear regression model 
# Calculate the correlation coefficient
rating = platforms_rating_df["rating"]
platforms = platforms_rating_df["number of platforms"]
correlation = st.pearsonr(rating,platforms)
print(f"The correlation ratings and number of platforms {round(correlation[0],2)}")
```

```python
#Calculate the linear regresssion 
slope, intercept, rvalue, pvalue, stderr = st.linregress(rating, platforms)
print("slope", slope)
print("intercept", intercept)
print("coefficient", rvalue)

rating_platforms_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(rating_platforms_eq)
rating_platforms_regression = rating * slope + intercept
rating_platforms_regression.head()
```

```python
# Plot the linear model on top of scatter plot 
plt.scatter(platforms_rating_df["rating"], platforms_rating_df["number of platforms"], facecolors="red", edgecolors="black") 
plt.plot(rating, rating_platforms_regression)

plt.title("Game ratings vs Average number of Platforms")
plt.xlabel("Ratings score")
plt.ylabel("Average number of platforms")
plt.grid()
plt.annotate(rating_platforms_eq,(1.5,9),fontsize=15,color="red")
# plt.xlim(14.5, 25.5)
# plt.ylim(33.5, 47)
```
