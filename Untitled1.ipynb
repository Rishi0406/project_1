```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
game_df=pd.read_csv("csvfiles/game.csv")
game_df["year"] = pd.DatetimeIndex(game_df["release date"]).year
years=game_df["year"].unique()
years=np.sort(years)
```

```python
year_df=pd.DataFrame()
for i in years:
    for j in range(1,13):
        year_df.loc[j,i]=0

year_df
```

```python
for i in game_df.index:
    for j in years:
        if game_df.loc[i,"year"]==j:
            month=game_df.loc[i,"month"]
            year_df.loc[month,j]=year_df.loc[month,j]+1
```

```python
year_df
```

```python
add_cat=["yet","owned","beaten","toplay","dropped","playing"]
for i in game_df.index:
    curr_added=game_df.loc[i,"added"].split("'")
    if "yet" in curr_added:
        ind=curr_added.index("yet")
        game_df.loc[i,"added_yet"]=curr_added[ind+1]
    if "owned" in curr_added:
        ind=curr_added.index("owned")
        game_df.loc[i,"added_owned"]=curr_added[ind+1]
    if "beaten" in curr_added:
        ind=curr_added.index("beaten")
        game_df.loc[i,"added_beaten"]=curr_added[ind+1]
```

```python
game_df=game_df.fillna(0)
game_df.head()
```

```python
for i in game_df.index:
    try:
        j=game_df.loc[i,"added_yet"].split(": ")[1]
        k=int(j.split(",")[0])
        game_df.loc[i,"added_yet"]=k
    except:
        continue

game_df["added_yet"].max()
```
