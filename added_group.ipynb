```python
# import dependencies
import pandas as pd
```

```python
# read game data
game_df=pd.read_csv("csvfiles/game.csv")
```

```python
# create and populate 'added' categories 
add_cat=["yet","owned","beaten","toplay","dropped","playing"]
for i in game_df.index:
    curr_added=game_df.loc[i,"added"].split("'")
    for j in add_cat:
        if j in curr_added:
            ind=curr_added.index(j)
            game_df.loc[i,j]=curr_added[ind+1]
```

```python
# remove nan and view dataframe
game_df=game_df.fillna(0)
game_df.head()
```

```python
# clean up added groups
for i in game_df.index:
    for j in add_cat:
        try:
            k=game_df.loc[i,j].split(": ")[1]
            try:
                k=int(k.split(",")[0])
            except:
                k=int(k.split("}")[0])
            game_df.loc[i,j]=k
        except:
            continue

game_df.head()
```

```python
# save dataframe
game_df.to_csv("csvfiles/game_addedgroup.csv",index=False)
```
